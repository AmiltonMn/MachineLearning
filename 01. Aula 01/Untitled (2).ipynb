```python
import pandas as pd

df = pd.read_csv('sample_data/california_housing_train.csv')
df = df.dropna()
df = df.fillna('Dado ausente')

latitudeMedia = df['longitude'].mean()
df['latitude'] = df['latitude'].fillna(latitudeMedia)
coluna_de_data = 'housing_median_age'

df[coluna_de_data] = pd.to_datetime(df[coluna_de_data])

for x in df.index:
  if df.loc [x, 'latitude'] < -200:
    df.loc[x, 'longitude'] = -144
  elif df.loc[x, 'longitude'] > 1000:
    df.drop.log[x, 'longitude']

df.drop_duplicates(inplace = True)
df
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['TipoSanguinio'] = le.fit_transform(df['TipoSanguinio'])
```
